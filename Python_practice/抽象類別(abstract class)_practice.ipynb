```python
import abc


class Car(abc.ABC):
    def __init__(self, name, color, power, publication_date):
        self.__name = name  # 車型品牌(名字)
        self.__color = color  # 車的顏色
        self.__power = power  # 車的馬力
        self.__publication_date = publication_date  # 出廠日期


    @abc.abstractmethod
    def detail(self):
        """
        output car's metadata
        """

    @property
    def name(self):
        return self.__name

    @property
    def color(self):
        return self.__color

    @property
    def power(self):
        return self.__power

    @property
    def publication_date(self):
        return self.__publication_date 
```

```python
class Super_car(Car):
    def __init__(self, name, color, power, publication_date, place, genre, price):
        super().__init__(name, color, power, publication_date)
        self.__place = place
        self.__genre = genre
        self.__price = price
    
    @property
    def price(self):
        return self.__price
    
    @property
    def place(self):
        return self.__place
    
    @property
    def genre(self):
        return self.__genre
    
    def detail(self):
        details = """
        車名 = {}
        顏色 = {}
        馬力 = {}
        出廠日期 = {}
        出廠地點 = {}
        類型 = {}
        價格 = {}
        """.format(self.name, 
                   self.color, 
                   self.power,
                   self.publication_date, 
                   self.place, 
                   self.genre, 
                   self.price)
        return details
```

```python
class Oil_car(Car):
    def __init__(self, name, color, power, publication_date, place, genre, price):
        super().__init__(name, color, power, publication_date)
        self.__place = place
        self.__genre = genre
        self.__price = price
    
    @property
    def price(self):
        return self.__price
    
    @property
    def place(self):
        return self.__place
    
    @property
    def genre(self):
        return self.__genre
    
    def detail(self):
        details = """
        車名 = {}
        顏色 = {}
        馬力 = {}
        出廠日期 = {}
        出廠地點 = {}
        類型 = {}
        價格 = {}
        """.format(self.name, 
                   self.color, 
                   self.power,
                   self.publication_date, 
                   self.place, 
                   self.genre, 
                   self.price)
        return details
```

```python
a = Super_car("Porsche-Cayman S", "紅色", "350hp", "2020/05/02", "台北", "跑車", "400萬")
print(a.detail())

b = Oil_car("BMW-M135i", "黑色", "320hp", "2021/04/05", "台北", "油車", "245萬")
print(b.detail())

c = EV_car("Tesla-Model 3", "白色", "300hp", "2021/06/05", "台北", "電動車", "160萬")
print(c.detail())

d = HEV_car("LEXUS-ct200h", "銀色", "136hp", "2021/02/23", "台北", "油電車", "140萬")
print(d.detail())
```
